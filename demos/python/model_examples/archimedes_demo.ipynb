# Archimedean Spiral Field

**-Reference:** Parker Archimedean spiral, see e.g. [Jokipii, Levy, Hubbard 1976](https://ui.adsabs.harvard.edu/abs/1977ApJ...213..861J/abstract),  [Jokipii, Levy, Hubbard 1977](https://ui.adsabs.harvard.edu/abs/1977ApJ...213L..85J/abstract)

**-Original Implementation:** [CRPropa](https://github.com/CRPropa/CRPropa3/tree/master)

**-Default parameters constrained using (see reference publication):** 

```python
import ImagineModels as img
from plotting_helpers import plot_slice

```

#### 1. Visualizing the model

```python
model = img.ArchimedeanMagneticField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmax = 10
vmin = -10
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=0, field_name='Archimedian spiral')
```

```python
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=1, field_name='Archimedian spiral')
```

```python
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=2, field_name='Archimedian spiral')
# disk field is zero in z-direction, but not halo, see below.
```

```python
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, field_name='Archimedian spiral', quiver=True, amplitude=True)
```

```python
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin, vmax, field_name='Archimedian spiral', amplitude=True)
```

```python
plot_slice(model_on_grid, 1, 1, shape, refpoint, increment, vmin, vmax, field_name='Archimedian spiral')
```

```python
plot_slice(model_on_grid, 2, 1, shape, refpoint, increment, vmin, vmax, field_name='Archimedian spiral')
# halo field is non-zero in z-direction!
```

```python
plot_slice(model_on_grid, 3, 1, shape, refpoint, increment, vmin, vmax, field_name='Archimedian spiral', quiver=True)
# field is in phi direction above plane, against it below plane, and has a z-component
```

```python
plot_slice(model_on_grid, 3, 0, shape, refpoint, increment, vmin, vmax, field_name='Archimedian spiral', quiver=True)
```
