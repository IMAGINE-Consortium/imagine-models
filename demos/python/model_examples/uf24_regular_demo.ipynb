#### 1. Visualizing the model

```python
model = img.UFMagneticField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -5
vmax = 5.

model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)

plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=0, field_name='UF24')



```

### 2. The parameters

## Changing the model class: 

This model has several variations. Per default, the parameters belonging to the "base" class are set. Additionally, the "neCL", "expX", "spur", "cre10", "synCG", "twistX" and "nebCor" variations are available, see the original publication referenced above for explanations

Additionally to changing the model parameters individually and manually (as explained in the general demo), the parameters here can be changed in bulk via the `set_parameters` function:



```python
print(model.activeModel)

model.set_parameters('neCL')

print(model.activeModel)

model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)

plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=0, field_name='UF24')
```
