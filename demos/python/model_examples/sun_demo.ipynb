# Sun et al. Field

**-Reference:** [Sun et al. (2007)](https://www.aanda.org/articles/aa/abs/2008/02/aa8671-07/aa8671-07.html) (ASS+Ring model implemented here) Updated parameters of the halo model were taken from [Sun et al. (2010)](https://iopscience.iop.org/article/10.1088/1674-4527/10/12/009)

**-Original Implementation:** [Hammurabi (old)](https://sourceforge.net/projects/hammurabicode/)

**-Default parameters constrained using (see reference publication):** all-sky maps in total intensity, linear polarization, and rotation measure (RM)

Here, we plot the Sun et al. magnetic field model and list its parameters.

```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.SunMagneticField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -3
vmax = 3
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 0, 2, shape, refpoint, increment, vmin, vmax, field_name='Sun et al. 2007')
```

```python
plot_slice(model_on_grid, 1, 2, shape, refpoint, increment, vmin, vmax, field_name='Sun et al. 2007')
```

```python
plot_slice(model_on_grid, 2, 2, shape, refpoint, increment, vmin, vmax, field_name='Sun et al. 2007')  # no z-component
```

```python
plot_slice(model_on_grid, 3, 2, shape, refpoint, increment, vmin, vmax, field_name='Sun et al. 2007', quiver=True)
```

```python
plot_slice(model_on_grid, 0, 1, shape, refpoint, increment, vmin, vmax, field_name='Sun et al. 2007')
```

```python
plot_slice(model_on_grid, 1, 1, shape, refpoint, increment, vmin, vmax, field_name='Sun et al. 2007')
```

```python
plot_slice(model_on_grid, 2, 1, shape, refpoint, increment, vmin, vmax, field_name='Sun et al. 2007')
```

```python
plot_slice(model_on_grid, 3, 1, shape, refpoint, increment, vmin, vmax, field_name='Sun et al. 2007')
# The strong halo field in the inner torus comes from the formulation of eq. 6 in the original publication. 
# It does not decrease with z, which is not physically expected.
```
