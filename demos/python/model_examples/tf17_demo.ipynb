# TF17 Field

**-Reference:** [Terral and Ferriere (2017)](https://arxiv.org/abs/1611.10222)

**-Original Implementation:** [CRPropa](https://github.com/CRPropa/CRPropa3/tree/master)

**-Default parameters constrained using (see reference publication):** Faraday rotation measures of extragalactic point sources, guided by radio polarization observations of external spiral galaxies, divB=0

Note that only the halo has been fitted using data, leading to very strong magnetic field strengths and unexpected features in the disk fields. Also, the halo fields can converge to infinite field strengths at infinite r/z. Thus, the field should be viewed more as a mathematical exercise than an actual model and should not be used for e.g. cosmic ray propagation.

```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.TFMagneticField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -100
vmax = 100.

# The model has different disk and halo fields, that can be changed in the following way:
model.activeDiskModel = 'Bd1'  # Ad1, Bd1 or Dd1
model.activeHaloModel = 'C1'   # C0 or C1
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=0, field_name='TF17')
```

```python
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=1, field_name='TF17')
```

```python
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=2, field_name='TF17')
# model has no z component
```

```python
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, field_name='TF17', quiver=True, amplitude=True)
```
