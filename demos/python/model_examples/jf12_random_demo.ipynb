```python
import ImagineModels as img
from plotting_helpers import plot_slice
import numpy as np
```

#### 1. Visualizing the model

```python
model = img.JF12RandomField()
shape = [100, 100, 20] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-10., -10., -2.]
vmin = -10
vmax = 10

seed = 23

model.spectral_offset = .1
model.spectral_slope = 3

model.apply_spectrum = True
model.clean_divergence = False

model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment, seed=seed)
print(model_on_grid[0].max())

profile = np.zeros(tuple(shape))

for i in range(shape[0]):
    for j in range(shape[1]):
        for k in range(shape[2]):
            profile[i, j, k] = model.spatial_profile(refpoint[0] + i*increment[0], 
                                                     refpoint[1] + j*increment[1],
                                                     refpoint[2] + k*increment[2])
plot_slice(profile, 2, shape, refpoint, increment, 0, 100)
plot_slice(profile, 1, shape, refpoint, increment, 0, 100)
plot_slice(profile, 0, shape, refpoint, increment, 0, 100)
```
