```python
import ImagineModels as img
from plotting_helpers import plot_slice
import numpy as np
```

#### 1. Visualizing the model

```python

# initializing and 
model = img.JF12RandomField()
shape = [100, 100, 20] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-10., -10., -2.]
vmin = -10
vmax = 10

seed = 23

# power spectrum parameters
model.apply_spectrum = True
model.spectral_offset = .1
model.spectral_slope = 3


model.clean_divergence = False

model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment, seed=seed)
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=0)
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=1)
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=2)

plot_slice(model_on_grid, 2, shape, refpoint, increment, -10, 10, amplitude=True, quiver=True)

plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin, vmax, vec_dim=0)
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin, vmax, vec_dim=1)
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin, vmax, vec_dim=2)
```

```python
profile = model.profile_on_grid(shape=shape, reference_point=refpoint, increment=increment)

plot_slice(profile, 2, shape, refpoint, increment, 0, 10, label=r"RMS amplitude / $\mu$G")
plot_slice(profile, 1, shape, refpoint, increment, 0, 10, label=r"RMS amplitude / $\mu$G")
plot_slice(profile, 0, shape, refpoint, increment, 0, 10, label=r"RMS amplitude / $\mu$G")
```

```python
random_numbers = model.random_numbers_on_grid(shape=shape, increment=increment, seed=seed)

plot_slice(random_numbers, 2, shape, refpoint, increment, -1, 1, label="")
plot_slice(random_numbers, 1, shape, refpoint, increment, -1, 1, label="")
plot_slice(random_numbers, 0, shape, refpoint, increment, -1, 1, label="")
```
