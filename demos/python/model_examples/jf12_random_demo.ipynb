#### 1. Visualizing the model

```python
model = img.JF12RandomField()
shape = [100, 100, 20] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-10., -10., -2.]
vmin = -50
vmax = 50

seed = 23

model.spectral_offset = .1
model.spectral_slope = 3

model.apply_spectrum = True
model.clean_divergence = False

model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment, seed=seed)
print(model_on_grid[0].max())

profile = model.profile_on_grid(shape=shape, reference_point=refpoint, increment=increment)

plot_slice(profile, 2, shape, refpoint, increment, 0, 10)
plot_slice(profile, 1, shape, refpoint, increment, 0, 10)
plot_slice(profile, 0, shape, refpoint, increment, 0, 10)
```
