# WMAP Field (Page et al 2007)

**-Reference:**  [Page et al. (2007)](https://iopscience.iop.org/article/10.1086/513699) 

**-Original Implementation:** [Hammurabi (old)](https://sourceforge.net/projects/hammurabicode/)

**-Default parameters constrained using (see reference publication):** WMAP emission (only intended as starting point, does not take into account any other data)


```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.WMAPMagneticField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -6.
vmax = 6.
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 0, 2, shape, refpoint, increment, vmin, vmax, field_name='WMAP 2007')
```

```python
plot_slice(model_on_grid, 1, 2, shape, refpoint, increment, vmin, vmax, field_name='WMAP 2007')
```

```python
plot_slice(model_on_grid, 2, 2, shape, refpoint, increment, vmin, vmax, field_name='WMAP 2007')
```

```python
plot_slice(model_on_grid, 3, 2, shape, refpoint, increment, vmin, vmax, field_name='WMAP 2007', quiver=True)
```

```python
plot_slice(model_on_grid, 0, 1, shape, refpoint, increment, vmin, vmax, field_name='WMAP 2007')
```

```python
plot_slice(model_on_grid, 1, 1, shape, refpoint, increment, vmin, vmax, field_name='WMAP 2007')
```

```python
plot_slice(model_on_grid, 2, 1, shape, refpoint, increment, vmin, vmax, field_name='WMAP 2007')
```
