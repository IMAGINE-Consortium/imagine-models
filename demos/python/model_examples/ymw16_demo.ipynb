#### 1. Visualizing the model

```python
model = img.UniformDensityField()
#model.do_loop = False
#model.do_gum = False
#model.do_galactic_center = False
#model.do_local_bubble = False
#model.do_spiral_arms = False
#model.do_thin_disc = True
#model.do_thick_disc = False
model.n0 = 1.5
shape = [10, 10, 5] # necessarily a list of 3 ints! 
increment = [.2, .2, .2]  
refpoint = [-1., -1., -.4]
vmin = 0
vmax = 1.e-6
grid_x = np.arange(20.) - 10
grid_y = np.arange(20.) - 10
grid_z = np.arange(20.) - 10
```

```python

model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
print(model_on_grid.shape)
print(model_on_grid.dtype)
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin=vmin, vmax=vmax, cut_index=None)
```

```python
for x in model_on_grid:
    for y in x:
        print(y)
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin=vmin, vmax=vmax, cut_index=None)
```
