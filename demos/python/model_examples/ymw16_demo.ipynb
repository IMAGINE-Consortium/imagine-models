```python
import ImagineModels as img
from plotting_helpers import plot_slice
import numpy as np
```

#### 1. Visualizing the model

```python
model = img.YMW16()
model.do_loop = True
model.do_local_bubble = True
model.do_gum = True
model.do_galactic_center = True
model.do_spiral_arms = True
model.do_thick_disc = True
model.do_thin_disc = True
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.15, .15, .1]  
refpoint = [-15., -15., -2.]
vmin = 0
vmax = 0.2
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin=0, vmax=vmax, cut_index=None, cmap="jet", plot_earth=False)
```

```python
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin=0, vmax=vmax, cut_index=None, cmap="jet", plot_earth=False)
```

```python
plot_slice(model_on_grid, 0, shape, refpoint, increment, vmin=0, vmax=vmax, cut_index=None, cmap="jet", plot_earth=False)
```
