# Stanev Field

**-Reference:** [Stanev (1997)](https://arxiv.org/abs/astro-ph/9607086) (bisymmetric model)

**-Original Implementation:** [Hammurabi (old)](https://sourceforge.net/projects/hammurabicode/)

**-Default parameters constrained using (see reference publication):** phenomenological arguments


```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.StanevBSSMagneticField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -6
vmax = 6.
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 0, 2, shape, refpoint, increment, vmin, vmax, field_name='Stanev 1997')
```

```python
plot_slice(model_on_grid, 1, 2, shape, refpoint, increment, vmin, vmax, field_name='Stanev 1997')
```

```python
plot_slice(model_on_grid, 2, 2, shape, refpoint, increment, vmin, vmax, field_name='Stanev 1997')
```

```python
plot_slice(model_on_grid, 3, 2, shape, refpoint, increment, vmin, vmax, quiver=True, field_name='Stanev 1997')
```

```python
plot_slice(model_on_grid, 0, 1, shape, refpoint, increment, vmin, vmax, field_name='Stanev 1997')
```

```python
plot_slice(model_on_grid, 1, 1, shape, refpoint, increment, vmin, vmax, field_name='Stanev 1997')
```
