# HMR Field

**-Reference:** [Harari, Mollerach, Roulet 1999](https://iopscience.iop.org/article/10.1088/1126-6708/1999/08/022)

**-Original Implementation:** [Hammurabi (old)](https://sourceforge.net/projects/hammurabicode/) and [Kachelrie√ü (2007)](https://arxiv.org/pdf/astro-ph/0510444.pdf ) 

**-Default parameters constrained using (see reference publication):** phenomenological arguments, used to study UHECR deflections, (de-)magnifications etc.

Here, we plot the Harari et al. magnetic field model and list its parameters.

```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.HMRMagneticField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -6
vmax = 6
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 0, 2, shape, refpoint, increment, vmin, vmax, field_name='HMR 1999')
```

```python
plot_slice(model_on_grid, 1, 2, shape, refpoint, increment, vmin, vmax, field_name='HMR 1999')
```

```python
plot_slice(model_on_grid, 2, 2, shape, refpoint, increment, vmin, vmax, field_name='HMR 1999')  # no field in z-direction
```

```python
plot_slice(model_on_grid, 3, 2, shape, refpoint, increment, vmin, vmax, quiver=True, field_name='HMR 1999')
```

```python
plot_slice(model_on_grid, 0, 1, shape, refpoint, increment, vmin, vmax, field_name='HMR 1999')
```

```python
plot_slice(model_on_grid, 1, 1, shape, refpoint, increment, vmin, vmax, field_name='HMR 1999')
```

```python
plot_slice(model_on_grid, 2, 1, shape, refpoint, increment, vmin, vmax, field_name='HMR 1999')
```

```python
plot_slice(model_on_grid, 3, 1, shape, refpoint, increment, vmin, vmax, field_name='HMR 1999')
```
