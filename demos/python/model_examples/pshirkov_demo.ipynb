# Pshirkov Magnetic Field (PT11)

**-Reference:** [Pshirkov, Tinyakov et al 2011](https://iopscience.iop.org/article/10.1088/0004-637X/738/2/192)

**-Original Implementation:** CRPropa3

**-Default parameters constrained using (see reference publication):** rotational measures from the NRAO VLA Sky Survey and a compilation by Kronberg and Newton-McGee

```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.PshirkovMagneticField()  # BSS field by default
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -5
vmax = 5
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 0, 2, shape, refpoint, increment, vmin, vmax)
```

```python
plot_slice(model_on_grid, 1, 2, shape, refpoint, increment, vmin, vmax)
```

```python
plot_slice(model_on_grid, 2, 2, shape, refpoint, increment, vmin, vmax)  # no field in z-direction, only in-plane
```

```python
plot_slice(model_on_grid, 3, 2, shape, refpoint, increment, vmin, vmax, quiver=True)
```

```python
plot_slice(model_on_grid, 0, 1, shape, refpoint, increment, vmin, vmax)
```

```python
plot_slice(model_on_grid, 1, 1, shape, refpoint, increment, vmin, vmax)
# from ref.: "the direction of the halo field is anticoincident with the direction of the disk field above the Galactic plane and coincident below"
```

```python
plot_slice(model_on_grid, 2, 1, shape, refpoint, increment, vmin, vmax)  # no field in z-direction
```
