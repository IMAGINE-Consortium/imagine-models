# Jansson Farrar Regular Field (JF12)

**-Reference:** [Jansson and Farrar 2012](https://ui.adsabs.harvard.edu/abs/2012ApJ...757...14J/abstract)

**-Original Implementation:** Hammurabi X

**-Default parameters constrained using (see reference publication)**: WMAP7 Galactic Polarized Synchrotron Emission (Stokes Q and U) maps from Gold et al. 2011; more than forty thousand extragalactic Faraday rotation measures


```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.JF12RegularField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -5
vmax = 5.

model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=0, field_name='JF12 regular')

```

```python
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, vec_dim=1, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 2, shape, refpoint, increment, -1, 1, vec_dim=2, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 2, shape, refpoint, increment, vmin, vmax, quiver=True, field_name='JF12 regular', amplitude=True)
# Note that in the center part < 5kpc we see the x-component of the field (explaining no arrows but larger field strengths)
```

```python
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin, vmax, vec_dim=0, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin, vmax, vec_dim=1, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin, vmax, vec_dim=2, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin, vmax, quiver=True, field_name='JF12 regular', amplitude=True)
# strong z-component (X-field) visible
```
