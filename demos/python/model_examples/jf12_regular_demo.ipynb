# Jansson Farrar Regular Field (JF12)

**-Reference:** [Jansson and Farrar 2012](https://ui.adsabs.harvard.edu/abs/2012ApJ...757...14J/abstract)

**-Original Implementation:** Hammurabi X

**-Default parameters constrained using (see reference publication)**: WMAP7 Galactic Polarized Synchrotron Emission (Stokes Q and U) maps from Gold et al. 2011; more than forty thousand extragalactic Faraday rotation measures

Here, we plot the Regular JF12 field and list its parameters.

```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.JF12RegularField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -3
vmax = 3.

model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 0, 2, shape, refpoint, increment, vmin, vmax, field_name='JF12 regular')

```

```python
plot_slice(model_on_grid, 1, 2, shape, refpoint, increment, vmin, vmax, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 2, 2, shape, refpoint, increment, -1, 1, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 3, 2, shape, refpoint, increment, 0, 5, quiver=True, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 0, 1, shape, refpoint, increment, vmin, vmax, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 1, 1, shape, refpoint, increment, vmin, vmax, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 2, 1, shape, refpoint, increment, vmin, vmax, field_name='JF12 regular')
```

```python
plot_slice(model_on_grid, 3, 1, shape, refpoint, increment, vmin, vmax, quiver=True, field_name='JF12 regular')
```
