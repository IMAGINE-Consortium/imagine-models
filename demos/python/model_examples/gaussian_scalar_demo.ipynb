```python
import ImagineModels as img
from plotting_helpers import plot_slice
import numpy as np
```

#### 1. Visualizing the model

```python
model = img.GaussianScalarField()
shape = [100, 100, 20] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-10., -10., -2.]
vmin = -1.
vmax = 1.

model.rms = 1.
model.spectral_offset = 0.01
model.spectral_slope = 2
model.apply_spectrum = True

seed = 23

model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment, seed=seed)
print(np.std(model_on_grid))
plot_slice(model_on_grid, 0, shape, refpoint, increment, vmin, vmax)

```

```python
plot_slice(model_on_grid, 1, shape, refpoint, increment, vmin, vmax)
```
