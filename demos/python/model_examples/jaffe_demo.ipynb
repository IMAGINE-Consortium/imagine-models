# Jaffe Field

**-Reference:**  [Jaffe et al. (2010)](https://ui.adsabs.harvard.edu/abs/2010MNRAS.401.1013J/abstract)

**-Original Implementation:** [Hammurabi X](https://github.com/hammurabi-dev/hammurabiX)

**-Default parameters constrained using (see reference publication):** total and polarised synchrotron emission data, rotation measures of extragalactic sources

```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.JaffeMagneticField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmax = 5.
vmin = -5.

# The model can be used either with a ring or bar feature (default) in the innermost region. These can be changed with:
model.ring = True
model.bar = False
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 0, 2, shape, refpoint, increment, vmin, vmax, field_name='Jaffe et al 2010')
```

```python
plot_slice(model_on_grid, 1, 2, shape, refpoint, increment, vmin, vmax, field_name='Jaffe et al 2010')
```

```python
plot_slice(model_on_grid, 2, 2, shape, refpoint, increment, vmin, vmax, field_name='Jaffe et al 2010')
# The model has no z-component
```

```python
plot_slice(model_on_grid, 3, 2, shape, refpoint, increment, vmin, vmax, field_name='Jaffe et al 2010', quiver=True)
```

```python
plot_slice(model_on_grid, 0, 1, shape, refpoint, increment, vmin, vmax, field_name='Jaffe et al 2010')
```

```python
plot_slice(model_on_grid, 1, 1, shape, refpoint, increment, vmin, vmax, field_name='Jaffe et al 2010')
```

```python
plot_slice(model_on_grid, 2, 1, shape, refpoint, increment, vmin, vmax, field_name='Jaffe et al 2010')
```

```python
plot_slice(model_on_grid, 3, 1, shape, refpoint, increment, vmin, vmax, field_name='Jaffe et al 2010')
```
