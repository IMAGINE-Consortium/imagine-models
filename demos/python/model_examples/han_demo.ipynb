```python
import ImagineModels as img
from plotting_helpers import plot_slice
```

#### 1. Visualizing the model

```python
model = img.HanMagneticField()
shape = [200, 200, 40] # necessarily a list of 3 ints! 
increment = [.20, .20, .20]  
refpoint = [-20., -20., -4.]
vmin = -3
vmax = 3.
```

```python
model_on_grid = model.on_grid(shape=shape, reference_point=refpoint, increment=increment)
plot_slice(model_on_grid, 0, 2, shape, refpoint, increment, vmin, vmax, field_name='Han et al 2018')
```

```python
plot_slice(model_on_grid, 1, 2, shape, refpoint, increment, vmin, vmax, field_name='Han et al 2018')
```
